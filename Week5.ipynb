```python
from nltk.corpus import state_union
import nltk
import matplotlib.pyplot as plt
import numpy as np
```

```python
men = []
women = []

for i, item in enumerate(state_union.fileids()):
    
    # Get words in lowercase
    words = state_union.words(item)
    words = [f.lower() for f in words]
    
    # Count apperaence of 'men' and 'women'
    men.append(words.count('men'))
    women.append(words.count('women'))
    
    #print(f'\'men\', {item}: {men[-1]}')
    #print(f'\'women\', {item}: {women[-1]}')


# Plotting
plt.plot(np.arange(len(state_union.fileids())), men)
plt.plot(np.arange(len(state_union.fileids())), women)
plt.show()


nltk.Text(state_union.words()).dispersion_plot(['men', 'women', 'man', 'woman'])
```

```python
def lexical_diversity(text, perc = False, tr = True):
    if perc:
        return percentage(len(set(text)), len(text), tr)
    return len(text) / len(set(text))

def percentage(count, total, tr):
    if tr:
        return count / total
    return 100 * count / total

{genre: lexical_diversity(brown.words(categories=genre), True) for genre in brown.categories()}
```
